Now make a separate list for numerical column so that it can be filled with mean

```python
col_list=[]
for i in test_csv:
    if (test_csv[i].dtypes=='float64') |(test_csv[i].dtypes=='int64')|(test_csv[i].dtypes=='int16'):
        col_list.append(i)
```

```python
col_list
```

```python
ts=test_csv[col_list]
tr=train_csv[col_list]
imputer=SimpleImputer(strategy='mean')
imputer.fit(tr)
x=imputer.transform(tr)
tst2=pd.DataFrame(x,columns=ts.columns)
trn2=pd.DataFrame(x,columns=tr.columns)
tst2.head()
#.describe()
```

```python
tst2[col_list].isnull().sum()
```

```python
trn2[col_list].isnull().sum()
```

Make every numerical coumn into int16 and merge to original dtaframe file

```python
for i in test_csv:
    if test_csv[i].dtypes=='float64':
        test_csv[i]=tst2[i].astype('int16')
        train_csv[i]=trn2[i].astype('int16')
    elif test_csv[i].dtypes=='int64':
        test_csv[i]=tst2[i].astype('int16')
        train_csv[i]=trn2[i].astype('int16')
    elif test_csv[i].dtypes=='object':
        test_csv[i]=test_csv[i].astype('category')
        train_csv[i]=train_csv[i].astype('category')
```
